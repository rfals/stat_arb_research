## Column Names

 - **open_time**: The time the bar opened
 - **open**: The price the bar opened at
 - **high**: The highest price the bar reached
 - **low**: The lowest price the bar reached
 - **close**: The price the bar closed at
 - **volume**: The volume of the bar
 - **close_time**: The time the bar closed
 - **quote_asset_volume**: The volume of the bar in the quote asset
 - **count**: The number of trades in the bar
 - **taker_buy_base_asset_volume**: The volume of the asset bought by takers
 - **taker_buy_quote_asset_volume**: The volume of the quote asset bought by takers
 - **ignore**: Ignore this column
